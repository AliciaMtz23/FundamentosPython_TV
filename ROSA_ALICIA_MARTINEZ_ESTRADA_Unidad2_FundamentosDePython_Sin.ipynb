## Ejecucin condicional: if, elif y else.

Como en cualquier lenguaje de programacin, las estructuras de control en Python permiten cambiar el flujo de ejecucin de los programas.

La palabra reservada *if* permite direccionar el flujo de ejecucin
del programa dependiendo de que alguna condicin puede ser
verdadera o falsa.

La sintaxis es la siguiente: 

**if** *condicin*:

> *accin1*
   
> *accion2*

> ...

> *accionM*

**Nota: la condicion del if en python no lleva parntesis** 

Como ejemplo prctico, hay que realizar un programa que resuelva ecuaciones de primer grado.

````python
a = float(input("Valor de a: "))
b = float(input("Valor de b: "))

x = -b/a

print("Solucion: ", x)
print("Hola Alicia.. ")
````

Para corregirlo se aadirn condiciones if para evaluar las posibles condiciones en las que puede incurrir el cdigo. Probarlo con valores positivos, negativos, a en cero y b en cero.

**NOTA:** Observe las tabulaciones que estn dentro de las condiciones if.

````python

a = float(input("Valor de a: "))
b = float(input("Valor de b: "))

if a != 0:
  x = -b / a
  print('Solucion: ', x)

if a == 0:
  if b != 0:
    print('La ecuacion no tiene solucion. ')
  if b == 0:
    print('La ecuacion tiene infinitas soluciones. ')
        
````

Si es necesario ejecutar acciones distintas cuando la evaluacin
de la condicin sea verdadera o falsa. Se puede utilizar la
palabra reservada *else*.

````python
a = float(input('Valor de a: '))
b = float(input('Valor de b: '))

if a == 0:
  if b == 0:
    print('La ecuacion tiene infinitas soluciones. ')
  else:
    print('La ecuacion no tiene solucion')  
````

Tambin es posible crear estructuras condicionales mltiples con la palabra reservada *elif*.

````python
a = 40
b = 41
if a < b:
  print(a, "Es menor que ", b)
elif a > b:
  print(a, "es mayor que ", b)
else: 
  print(a, "Es igual que ", b)    
````

### Operador ternario.

En Python existen tambin el operador ternario, se trata de una clsula *if - else* que se define dentro de una sola lnea, incluso puede ser utilizada dentro de un print(). La estructura es la siguiente:

*`[cdigo si se cumple] if [condicin] else [cdigo si no se cumple]`*



````python
x = 8
print("Es 5" if x == 5 else "No es 5")
````

Tambin se puede utilizar para asignar valores a variables:

````python
a = 10
b = 5
c = a/b if b!=0 else -1
print(c)
````

## Iteraciones: while, for.


### Ciclo for.

Este ciclo se caracteriza por tener un nmero de iteraciones definido de antemano. No tiene un candicin de paro , ms bien contiene un *iterable* que define las veces que se ejecutar el cdigo. El siguiente ejemplo representa un ciclo representa un contador:

````python
for cont in range(0, 6, 1):
  print("Valor de contador: ", str(cont))
````

### Iterables e iteradores

Los **iterables** son aquellos objetos que como su nombre indica pueden ser iterados, lo que dicho de otra forma es, que puedan ser indexados. Si piensas en una`list`, podemos indexarlo con `lista[1]` por ejemplo, por lo que sera un iterable. Algunos ejemplos son: listas, tuplas, cadenas o diccionarios.

Los **iteradores** son objetos que hacen referencia a un elemento, y que tienen un mtodo `next` que permite hacer referencia al siguiente.

El ciclo for es capaz de recorrer los objetos iterables mediante la siguiente sintaxis:

```
for <variable> in <iterable>:
    <Cdigo>
```

Para saber si un objeto es iterable, se puede usar la funcin `isinstance()`:

````python
from collections.abc import Iterable

lista = [1, 2, 3, 4]
cadena = "python"
numero = 10
myset = {"Platano", "Manzana", "Cereza"}

print(isinstance(lista, Iterable))
print(isinstance(cadena, Iterable))
print(isinstance(numero, Iterable))
print(isinstance(myset, Iterable))
````

La funcin `iter()` puede ser llamada desde un objeto que sea iterable y retorna un iterador, que es una variable que hace referencia al objeto iterable original y permite acceder a sus elementos con `next()`. Cuando coimienza, el iterador apunta afuera de la lista, y no hace referencia al primer elemento hasta que se llema a `next()` por primera vez.

````python
it = iter(lista)
print(next(it))
print(next(it))
print(next(it))
````

Existen iteradores para diferentes clases:

`str_iterator` para cadenas.

`list_iterator` para sets.

`tuple_iterator` para tuplas.

`set_iterator` para sets.

`dict_keyiterator` para diccionarios.

El ciclo for, como ya se mencion anteriormente, permite recorrer objetos iterables, ejemplos:

````python
for i in cadena:
  print(i)
````

````python
for i in myset:
  print(i)
````

La funcin `range()` genera una secuencia de nmeros que van desde 0 por defecto hasta el nmero que se pasa como parmetro menos 1. En realidad, se pueden pasar hasta tres parmetros separados por coma, donde el primer es el inicio de la secuencia, el segundo el final y el tercero el salto que se desea entre nmeros. Por defecto se empieza en 0 y el salto es de 1:

`range(inicio, fin, salto)`

````python
for i in range (5, 20, 2):
  print(i)
````

````python
for i in range(5, 0 , -1):
  print(i)
````

Los ciclos `for` tambin pueden anidarse, como en cualquier lenguaje de programacin:

````python
lista = [[56, 34, 1],
         [12, 4, 5],
         [9, 4, 3]]
for i in lista:
  print(i)

print("\nFor anidado\n") 

for i in lista:
  for j in i:
    print(j)
````

## Ciclo while

EL ciclo while repetir las sentencias dentro de l, mientras se cumpla la condicin especificada.

````python
 x = 0
 while x < 6:
   print("Valor de contador: ", x)
   x += 1

print("Fin del ciclo")   
````

Tambin es posible escribir ciclos while en una sla lnea.

````python
x = 6
while x > 0: print("Valor de contador: ", x); x -= 1

print("Fin del ciclo")
````

Tambin permite recorrer algn elemento iterable mientras existan elementos en l, por ejemplo, una lista y en cada iteracin se va eliminando un dato:

````python
x = ["uno", "dos", "tres"]
while x: 
  x.pop(0)
  print(x)

  
````

En Python, es posible mezclar la palabra reservada `else`, con el `while`. Servir para ejecutar cdigo una vez que el ciclo all terminado normalmente:

````python
x = 0
while x < 6:
  print("Valor del contador", x)
  x += 1
else:
  print("Fin del ciclo")
 
````

Como el ciclo `while` ejecuta las sentencias en su interior cuando la condicin es verdadera, se deber tener cuidado de no generar ciclos infinitos, a menos que sea requerido por el programa a realizar. El siguiente ejemplo es vlido, pero ocasionar que el programa qued "colgado" infinitamente:
```
while True:

```

## Modificacin de ciclos.

`Break` permite detener la ejecucin del ciclo. Una vez que el interprete encuentre la palabra, el ciclo terminar, sin embargo se puede considerar que es un final forzado.

Se puede utilizar tanto con `ciclos` for y `while`. Comnmente se usa para establecer una condicin de parada del ciclo y as evitar consumir ciclos de reloj de la computadora:

````python
for letra in cadena:
  if letra == 'h':
    print("Se encontro la h")
    break
  print(letra)  
````

El siguiente ejemplo muestra como un ciclo infinito se puede detener con una condicin y un sentencia `break`:

````python
x = 5
while True:
  x-= 1
  print(x)

  if x == 0:
    break

print("Fin del ciclo")
````

Si `break` se encuentra dentro de ciclos anidados, nicamente detenda el ciclo en el que se encuentra:

````python
for i in range(0, 4):
  for j in range(0, 4):
    break
    print(i,j)
````

La otra sentencia que permite modificar los ciclos en Python es la palabra reservada `continue`. sta permite que dentro de un ciclo se "salte" las lneas posteriores a donde se encuentra, haciendo que el ciclo continue su ejecucin en la siguiente iteracin:

````python
for letra in cadena:
  if letra  == 't':
    continue
  print(letra)  
````

Como se puede observar, si bien se brinco el `print()`, nicamente no lo ejecuta en esa iteracin, siendo que en las dems se ejecuta normalmente.

Un ejemplo con `while`:

````python
x = 5
while x > 0:
  x-= 1
  if x == 3:
    continue
  print(x)  
````
